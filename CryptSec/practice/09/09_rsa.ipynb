# RSA

**Emlék**: (Kis Fermat-tétel): $$ a^{p-1} \equiv \left\{ \begin{array}{rcl}
1 \pmod{p}, & \text{ha} & p\nmid a \\ 
0 \pmod{p}, & \text{ha} & p \mid a \\
\end{array}\right.$$

Igaz-e a kis Fermat-tétel, ha $p$-t kicseréljük egy olyan számra, ami nem prímszám? Legyen $m=15485207$:



Akkor nem igaz ha p nem prim. Lehet hogy igaz, de nem garantalt a jol mukodes.


De a kis Fermat-tétel általánosítható. Legyen $m = 15$, ekkor 
$$
\begin{array}{ll}
    a^4 \equiv 1 \pmod{15} & \text{ha}\; a=1,2,4,7,8,11,13,14 \\
    a^4 \not\equiv 1 \pmod{15} & \text{ha}\; a=3,5,6,9,10,12 \\
\end{array}
$$
**Kérdés**: Mi különbözteti meg $a=1,2,4,7,8,11,13,14$ számokat az $a=3,5,6,9,10,12$ számoktól? 

**Megoldás**: Az első lista elemei relatív prímek $15$-tel, míg a második lista elemeinek a prímtényezős felbontása tartalmaz olyan számokat, amiket $15$ felbontása is tartalmaz. Azaz ha $a$ és az $m$ modulus relatív prímek, akkor a kis Fermat-tétel *valamilyen* formája igaz kell, hogy legyen. De ekkor a helyes kitevő nem feltétlenül $m - 1$.

A $4$ mint kitevő miért jó? Nézzük meg az alábbi kongruenciákat:
$$
a^4 \equiv 1 \pmod{3}, \quad a^4 \equiv 1 \pmod{5}
$$
Ebből a két kongruenciából kijön, hogy $a^4 \equiv 1 \pmod{15}$. (A két kongruencia igaz, ami a kis Fermat-tételből kijön)

Még egy észrevétel: a $4$, mint kitevő, $p-1$ többszöröse $p=3$ és $p=5$ esetén *is*.

**Tétel** \(Euler\-formula $pq$\-ra\): Legyenek $p,q$ különböző prímszámok és legyen $g = \text{gcd}(p-1, q-1)$. Ekkor $$a^{(p-1)(q-1)/g} \equiv 1 \pmod{pq}$$ minden olyan $a$ számra, amire $\text{gcd}(a,pq)=1$ \(relativ primek\) teljesül. Ha $p,q$ páratlan prímek, akkor $$a^{(p-1)(q-1)/2} \equiv 1 \pmod{pq}$$ minden olyan $a$ számra, amire $\text{gcd}(a,pq)=1$ \(relativ primek\) teljesül.


Diffie-Hellman: $a^x \equiv b \pmod{p}$ nehézségére épít

RSA: $x^e \equiv c \pmod{N}$ nehézségére épít (azaz nehéz $e$-edik hatványt kiszámolni modulo $N$). Ez mennyire reális, ha $N$ prímszám?

**Tétel**: Legyen $p$ prímszám és $e \geq 1$ úgy, hogy $\text{gcd}(e, p-1) = 1$. Ekkor $e$-nek létezik multiplikatív inverze modulo $p - 1$, $d$, azaz $de \equiv 1 \pmod{p-1}$. Ekkor az $x^e \equiv c \pmod{p}$ kongruenciának $x \equiv c^d \pmod{p}$ megoldása lesz és az $x$ egyértelmű.

**Példa**: Legyen $m = 7919, e = 1583, c = 4714$. Oldjuk meg az $x^{1583} \equiv 4714 \pmod{7919}$ kongruenciát!

Ha $N=pq$, akkor már más a helyzet:

**Tétel**: Legyenek $p,q$ prímszámok és $e \geq 1$ úgy, hogy $\text{gcd}(e, (p-1)(q-1)) = 1$. Ekkor $e$-nek létezik multiplikatív inverze modulo $(p-1)(q-1)$, azaz $de \equiv 1 \pmod{(p-1)(q-1)}$. Ekkor az $x^e \equiv c \pmod{pq}$ kongruenciának $x \equiv c^d \pmod{pq}$ megoldása lesz és az $x$ egyértelmű. 

**Megjegyzés**: Ha nem ismert $N$ felbontása, akkor a fenti tételt nem lehet használni a kongruencia megoldására

**Példa**: Legyen $N = 30069476293, e = 9843, c = 134872$. Oldjuk meg az $x^{9843} \equiv 1348772 \pmod{30069476293}$ kongruenciát!

## (Textbook) RSA (Rivest-Shamir-Adleman)

Kulcsgenerálás:
- Alice választ két prímet, amit titokban tart, de $N=pq$ értékét nyilvánossá teszi

Titkosítás:
- Alice választ egy *nyilvános* $e$ kitevőt úgy, hogy $\text{gcd}(e, (p-1)(q-1))=1$ teljesüljön
- A titkosítandó $m$ üzenet egy egész szám lesz $1 \leq m < N$ között
- Bob titkosítja az $m$ üzenetet $e,N$ felhasználásával: $c \equiv m^e \pmod{N}$
- Bob elküldi Alice-nak a titkosított üzenetet

Visszafejtés:
- Alice ismeri $p,q$ értékét, ezért ki tudja számolni $e$ multiplikatív inverzét: $de \equiv 1 \pmod{(p-1)(q-1)}$
- A $d$ értékét felhasználva vissza tudja fejteni az üzenetet: $m \equiv c^d \pmod{N}$

**Megjegyzések**:
- Választhatjuk $e=3$ értéket is, *elvileg* ez ugyanolyan biztonságos, mint valami nagyobb érték ($e=1$ miért nem jó?)
- Gyakori választás $e=2^{16}+1 = 65537$ - ezzel a titkosítás gyors marad (4 hatványozás és 1 szorzással kiszámolható $m^e \mod{N}$)
- De $d > N^{1/4}$ kell legyen

Egy támadás:

- Eve megkéri Alice-t, hogy fejtsen vissza egy üzenetet, Alice örömmel megteszi (így authentikája magát)
- Eve elkapja Bob $c$ ciphertext-jét.
- Eve választ egy random $k$ értéket és Alice-nak továbbítja a $c' \equiv k^e \cdot c \pmod{N}$ ciphertext-et
- Alice visszafejti $c'$-t és elküldi Eve-nek: $m' \equiv (c')^d \equiv (k^e \cdot c)^d \equiv (k^e \cdot m^e)^d \equiv k \cdot m \pmod{N}$
- Eve ebből megismeri $k \cdot m \pmod{N}$ értékét, és mivel $k$-t ő választotta, megkapja Bob $m$ üzenetét

Megjegyzések:

1. Eve visszafejtette Bob üzenetét *anélkül*, hogy faktorizálná $N$-et
2. Alice nem fogja felismerni, hogy Eve egy támadó lenne, mert $k^e \cdot c \pmod{N}$ és $k \cdot m \pmod{N}$ véletlenszerűnek tűnik

Ugyanitt: Egy $N$-hez csak egy $e$ kitevőt választunk:

- Válasszunk $e_1,e_2$ kitevőket
- Bob egy $m$ üzenetet a két kitevővel titkosít: $c_1 \equiv m^{e_1} \pmod{N}$ és $c_2 \equiv m^{e_2} \pmod{N}$
- Eve ezt lehallgatja, amiből kap egy megoldást $e_1 \cdot u + e_2 \cdot v = \text{gcd}(e_1, e_2)$\-re
- Ebből vissza tudja fejteni az üzenetet, ha $\text{gcd}(e_1, e_2) = 1$: $c_1^u \cdot c_2^v \equiv (m^{e_1})^u \cdot (m^{e_2})^v \equiv m^{e_1 \cdot u + e_2 \cdot v} \equiv m^{\text{gcd}(e_1,e_2)} \pmod{N}$
