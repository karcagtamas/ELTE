Jelölések, paraméterek:
- kulcstér: $\mathcal{K}$ a lehetséges kulcsok halmaza, valószínűségi algoritmus
- üzenettér: $\mathcal{M}$ a lehetséges üzenetek halmaza
- titkos üzenettér: $\mathcal{C}$ a lehetséges titkos üzenetek halmaza
- a $\mathcal{K}$ és $\mathcal{C}$ véges halmazok
- megengedjük, hogy $\mathtt{Enc}$ valószínűségi algoritmus legyen
- ugyanúgy elvárjuk, hogy $\mathtt{Dec}_k(\mathtt{Enc}_k(m)) = m$

Egyéb definíciók:
- Legyen $K$ egy valószínűségi változó $\mathcal{K}$ felett: $Pr(K = k)$ annak a valószínűsége, hogy a $\mathtt{Gen}$ algoritmus a $k \in \mathcal{K}$ kulcsot választotta
- Legyen $M$ egy valószínűségi változó $\mathcal{M}$ felett: $Pr(M = m)$ annak a valószínűsége, hogy az $m \in \mathcal{M}$ üzenetet titkosítjuk
- Felteszzük, hogy $M$ és $K$ függetlenek

**Példa**: Legyen $\mathcal{K} = \{0, \dots, 25\}$ és $Pr(K = k) = 1/26$ minden $k \in \mathcal{K}$-ra. Eltolásos titkosítást alkalmazunk. Tegyük fel, hogy a lehetséges üzenetek eloszlása a következő: $Pr(M = \mathtt{a}) = 0.7$ és $Pr(M = \mathtt{z}) = 0.3$.
1. Mi a valószínűsége, hogy a $\mathtt{B}$ titkosított üzenetet kapjuk?
2. Mi annak a valószínűsége, hogy az $\mathtt{a}$ üzenetet titkosítottuk, feltéve, hogy a $\mathtt{B}$ volt a titkosított üzenet?

**Definíció** (Tökéletes biztonság): Egy $(\mathtt{Gen}, \mathtt{Enc}, \mathtt{Dec})$ titkosítási az $\mathcal{M}$ üzenettérrel tökéletesen biztonságos, ha minden $M$ feletti eloszlásra, minden $m \in \mathcal{M}$ üzenetre és minden olyan $c \in \mathcal{C}$ titkos üzenetre, amire $Pr(C = c) > 0$: $$Pr(M = m\,|\,C = c) = Pr(M = m)$$

## A biztonság alternatív definíciója

(Ez ekvivalens az előző definícióval)

Legyen $\Pi = (\mathtt{Gen}, \mathtt{Enc}, \mathtt{Dec})$ egy titkosítási séma az $\mathcal{M}$ üzenettérrel. Legyen $\mathcal{A}$ egy támadó algoritmus. A lehallgatásos megkülönböztethetetlenség kísérlet $\mathtt{PrivK}^{\text{eav}}_{\mathcal{A},\Pi}$:
1. az $\mathcal{A}$ támadó kiad $m_0,m_1$ üzeneteket, amire $|m_0| = |m_1|$ és $m_0, m_1 \in \mathcal{M}$
2. a $k$ kulcsot a $\mathtt{Gen}$ algoritmussal véletlenszerűen generáljuk, és választunk $b \in_R \{0,1\}$ bitet. Legyen $c = \mathtt{Enc}_k(m_b)$, amit átadunk $\mathcal{A}$-nak.
3. az $\mathcal{A}$ támadó kiad egy $b' \in \{0,1\}$ bitet
4. Ha $b' = b$, akkor $\mathtt{PrivK}^{\text{eav}}_{\mathcal{A},\Pi} = 1$ és a kísérlet sikerek, egyébként $\mathtt{PrivK}^{\text{eav}}_{\mathcal{A},\Pi} = 0$ és a kísérlet sikertelen.

**Tétel**: Egy $\Pi$ titkosítási séma tökéletesen biztonságos $\mathcal{M}$ felett, ha $\forall \mathcal{A}$ támadóra $$Pr(\mathtt{PrivK}^{\text{eav}}_{\mathcal{A},\Pi} = 1) = \frac{1}{2}.$$

**Feladat**: Legyen $\mathcal{M} = \{a,b\}^2$ és $t \in \{1, 2\}$. Mutassuk meg, hogy a Vigenère-féle titkosításra *nem* teljesül a fenti definíció, azaz adjunk meg egy olyan támadót, amire $Pr(\mathtt{PrivK}^{\text{eav}}_{\mathcal{A},\Pi} = 1) > \frac{1}{2}.$

## One-Time Pad

Gilbert Vernam, 1926, de korábban már létezett.

Legyen $n > 0$ rögzített. Ekkor $\mathcal{K} = \mathcal{M} = \mathcal{C} = \{0,1\}^n$.
- $\mathtt{Gen}$: egyenletes eloszlás alapján választ egy $k \in \mathcal{K}$ kulcsot
- $\mathtt{Enc}$: adott $k \in \{0, 1\}^n$ kulcsra és $m \in \{0,1\}^n$ üzenetre $c := k \oplus m$
- $\mathtt{Dec}$: adott $k \in \{0, 1\}^n$ kulcsra és $c \in \{0,1\}^n$ titkos üzenetre $m := k \oplus c$

## *Titkosítási sémák algebrája

Claude Shannon (1949)

**Definíció**: Legyen $\mathcal{M}$ egy üzenettér, $\mathcal{C}$ és $\mathcal{C}'$ két titkos üzenet-tér. Legyenek $\mathbb{S}_1$ és $\mathbb{S}_2$ két titkosítási séma úgy, hogy $$\mathtt{Enc}_{\mathbb{S}_1} : \mathcal{M} \rightarrow \mathcal{C}\quad \text{és}\quad \mathtt{Enc}_{\mathbb{S}_2} : \mathcal{C} \rightarrow \mathcal{C}'.$$ Ekkor $\mathbb{S} = \mathbb{S}_1 \times \mathbb{S}_2$ egy új titkosítási séma lesz és $\mathtt{Enc}_{\mathbb{S}} : \mathcal{M} \rightarrow \mathcal{C}'$. A titkosítás során előbb $\mathtt{Enc}_{\mathbb{S}_1}$-et, majd $\mathtt{Enc}_{\mathbb{S}_2}$-t használjuk.

Ez mindig asszociatív: $(\mathbb{S}_1 \times \mathbb{S}_2) \times \mathbb{S}_3 = \mathbb{S}_1 \times (\mathbb{S}_2 \times \mathbb{S}_3)$
