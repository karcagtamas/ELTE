## Titkosítási sémák

Egy titkosítási séma:
- A `Gen` *kulcsgeneráló algoritmus* egy valószínűségi algoritmus, ami egy $k$ kulcsot generál (általában) egyenletes eloszlás szerint.
- Az `Enc` *titkosító algoritmus* egy $k$ kulcsot és $m$ üzenetet (plaintext) vár benemetként, amikből előállítja a $c$ titkosított üzenetet (ciphertext). Jelölés: $\mathtt{Enc}_k(m)$.
- A `Dec` *visszafejtő algoritmus* egy $k$ és $c$ titkosított üzenetet vár bemenetként, amiből előállítja az $m$ üzenetet. Jelölés: $\mathtt{Dec}_k(c)$.

Elvárás: $\mathtt{Dec}_k(\mathtt{Enc}_k(m)) = m$ minden $k$ kulcsra és $m$ üzenetre

Formálisan:

- $\mathtt{Enc} : \mathcal{K} \times \mathcal{M} \rightarrow \mathcal{C}$
- $\mathtt{Dec} : \mathcal{K} \times \mathcal{C} \rightarrow \mathcal{M}$

A titkosítandó üzenetet kisbetűkkel, a titkosított üzenetet nagybetűkkel jelöljük
