```python
def keygen():
    res = {}
    perm = random.sample(ascii_uppercase, k=26)
    for k, v in zip(ascii_lowercase, perm):
        res[k] = v
    return res

def keygen():
    return dict(zip(ascii_lowercase, random.sample(ascii_uppercase, k=26)))
```

```python
keygen()
```

```python
def encrypt(key, plaintext):
    res = ''
    for p in plaintext:
        res += key[p]
    return res


def decrypt(key, ciphertext):
    key = {v: k for k, v in key.items()}
    return encrypt(key, ciphertext)
```

```python
key = keygen()
decrypt(key, encrypt(key, plaintext)) == plaintext
```

## Vigenère-féle titkosítás

Blaise de Vigenère (1586): *Traicté des Chiffres*

Titkosíott szöveg statisztikai jellemzőinek elkenése egy kis trükkel.

```python
def repeat_to_length(s, length):
    return (s * (length // len(s) + 1))[:length]
```

A kulcs tetszőlegesen választható:

```python
key = 'kulcs'
plaintext = "ezsembiztonsagos"
```

```python
key = repeat_to_length(key, len(plaintext))
print(key)
print(plaintext)
```

```python
def encrypt(key, plaintext):
    c = ''
    for k, p in zip(key, plaintext):
        k_ind = ascii_lowercase.find(k)
        c += ascii_uppercase[(ascii_lowercase.find(p) + k_ind) % 26]
    return c
```

```python
def decrypt(key, ciphertext):
    p = ''
    for k, c in zip(key, ciphertext):
        k_ind = ascii_lowercase.find(k)
        p += ascii_lowercase[(ascii_uppercase.find(c) - k_ind) % 26]
    return p
```

```python
decrypt(key, encrypt(key, plaintext)) == plaintext
```

Hogyan tovább? Ha $t$ a kulcs hossza (a fenti példában $5$), akkor hány lehetséges kulcs van? Tudunk-e jobbat?

Mennyit javítottunk ezzel?

Második módszer: William Friedman, 1920

*Index of coincidence*: Legyen $\mathbf{x} = x_1 x_2 \cdots x_n$ egy sztring. Az **x** *koincidencia indexe* $I_c(\mathbf{x})$ annak a valószínűsége, hogy $\mathbf{x}$ két random eleme megegyezik.

Legyen $n$ az $\mathbf{x}$ szöveg hossza és jelölje $f_i$ az $i$ karakter előfordulásainak számát $\mathbf{x}$-ben. Ekkor az $\mathbf{x}$ koincidencia indexe: $$I_c(\mathbf{x}) = \frac{\sum_{i=0}^{25}\binom{f_i}{2}}{\binom{n}{2}} = \frac{1}{n(n-1)}\sum_{i=0}^{25}f_i(f_i - 1)$$

```python
def index_of_coincidence(x):
    n = len(x)
    obs = Counter(x)
    idc = 0
    for c, fi in obs.items():
        idc += fi*(fi - 1) / (n * (n-1))
    return idc
```

```python
from math import comb

def index_of_coincidence2(x):
    obs = Counter(x)
    return sum(comb(fi, 2) for fi in obs.values()) / comb(len(x), 2)
```

```python
index_of_coincidence2('thiscryptosystemisnotsecure')
```

```python
index_of_coincidence2(random.choices(ascii_lowercase, k=10000))
```

Fontos észrevétel: Legyen $\mathbf{x}$ egy értelmes angol sztring. Ekkor $I_c(\mathbf{x}) \approx \sum_{i=0}^{25}p_i^2 \approx 0.065$

### A törés lépései:

1. Keressük meg a kulcs hosszát, $t$-t: osszuk fel a `ciphertext`-et $t$ részre!
Az eredeti ciphertext: $c = c_1 c_2 \cdots c_n$
A felosztás után *t* darab részsztringet kapunk:
\begin{align*}
c_1 &= c_1 c_{t+1} c_{2t+1} \cdots, \\
c_2 &= c_2 c_{t+2} c_{2t+2} \cdots, \\
\vdots \\
c_t &= c_t c_{2t} c_{3t} \cdots
\end{align*}
Állítás: ha $t$ a kulcs hossza, akkor $\forall c_i: I_c(c_i) \approx 0.065$.

```python
ciphertext[:10]  # első 10 karakter
```

```python
ciphertext[-10:]  # utolsó 10 karakter
```

```python
ciphertext[::2]  # minden második karakter
```

```python
t = 1
while True:
    parts = [index_of_coincidence2(ciphertext[s::t]) for s in range(t)]
    if all([.055 <= p <= .075 for p in parts]):
        break
    t += 1
t
```

2. Keressük meg a titkos kulcsot. Minden $c_j$ ($1 \leq j \leq t$) részsztringet toljunk el $0 \leq g \leq 25$-tel. Definiáljuk a következő mennyiséget: $$M_g = \sum_{i=0}^{25} p_i\frac{f_{i + g}}{n'}$$ ahol $n'$ a $c_j$ hossza. Ha $g = k_i$, akkor $M_g \approx \sum_{i=0}^{25} \approx 0.065$. 

```python
possible_key = ''

for c in range(t):
    cj = ciphertext[c::t]
    obs = Counter(cj)
    max_idc = (0, "")
    for g in range(26):
        idc = 0
        for i in range(26):
            idc += freq[ascii_lowercase[i]] * obs[ascii_uppercase[(i + g) % 26]] / len(cj)
        if .055 <= idc <= .075 and max_idc[0] < idc:
            max_idc = (idc, ascii_lowercase[g])
    possible_key += max_idc[1]
possible_key
```

```python
decrypt(repeat_to_length(possible_key, len(ciphertext)), ciphertext)
```

**Állítás**: A Vigenere-féle titkosítás biztonságos, ha a kulcs hossza a titkosítandó szöveg hosszával egyezik meg.
