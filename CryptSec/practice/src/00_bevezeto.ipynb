## A Jupyter notebook és főbb parancsok

A notebook egy webalkalmazás, aminek a célja, hogy kódot osszunk meg egymással magyarázó szöveggel együtt, így segítve egy probléma, illetve annak megoldásának a bemutatását. 

A notebook cellákra oszlik, ahol lehet Markdown formázott szöveget, LaTex egyenleteket, illetve kódot írni. Egy notebook egy kernel-t tud futtatni, ami tulajdonképpen a kiválasztott programozási nyelv integrálva a Jupyterbe (főként interpretált nyelvek). A Python-t kivéve jellemzően ezeket telepíteni kell. A [Jupyter wiki](https://github.com/jupyter/jupyter/wiki) hasznos forrás.

Főbb billentyűkombinációk:

- új cella létrehozása: `a` és `b`, a kiválasztott cella alá, illetve fölé
- cella módosítása: `Enter` a kiválasztott cellára, az `Esc` kilép a módosításból
- egy cella futtatása: `Ctrl+Enter` vagy `Shift+Enter`. A különbség, hogy a `Ctrl+Enter` a lefuttatott cellán marad, míg a `Shift+Enter` a következő cellára ugrik (vagy létrehoz egy újat, ha nincs következő cella)
- cella törlése: `dd`

Egy cella típusa lehet:

- Markdown formázott: `m` (korlátozottan tud LaTeX-et is)
- kód: `y`

További parancsok és kombinációk a menüben, de ezek a legfontosabbak.

## Titkosítási sémák

Egy titkosítási séma:
- A `Gen` *kulcsgeneráló algoritmus* egy valószínűségi algoritmus, ami egy $k$ kulcsot generál (általában) egyenletes eloszlás szerint.
- Az `Enc` *titkosító algoritmus* egy $k$ kulcsot és $m$ üzenetet (plaintext) vár benemetként, amikből előállítja a $c$ titkosított üzenetet (ciphertext). Jelölés: $\mathtt{Enc}_k(m)$.
- A `Dec` *visszafejtő algoritmus* egy $k$ és $c$ titkosított üzenetet vár bemenetként, amiből előállítja az $m$ üzenetet. Jelölés: $\mathtt{Dec}_k(c)$.

Elvárás: $\mathtt{Dec}_k(\mathtt{Enc}_k(m)) = m$ minden $k$ kulcsra és $m$ üzenetre

Formálisan:

- $\mathtt{Enc} : \mathcal{K} \times \mathcal{M} \rightarrow \mathcal{C}$
- $\mathtt{Dec} : \mathcal{K} \times \mathcal{C} \rightarrow \mathcal{M}$

A titkosítandó üzenetet kisbetűkkel, a titkosított üzenetet nagybetűkkel jelöljük
