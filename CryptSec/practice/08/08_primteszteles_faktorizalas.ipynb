**Kérdés**: Ez bizonyítja, hogy a fenti $n$ prímszám?

Ha $n$ összetett szám, akkor $2^n \not\equiv 2 \pmod{n}$, egyébként *valószínűleg* prímszám. Ehhez kapcsolódik a következő definíció:

**Definíció**: Legyen $n \in \mathbb{Z}$ rögzített egész szám. Azt mondjuk, hogy az $a$ egész szám az $n$ egy tanúja (hogy $n$ összetett), ha $a^n \not\equiv a \pmod{n}$

```python
def is_composite(n):
    z = Zmod(n)
    return any([z(i) ^ n != z(i) for i in range(n)])
```

```python
is_composite(137), is_composite(291), is_composite(561)
```

## Miller-Rabin teszt

**Tétel**: Legyen $p$ egy páratlan prímszám és $p-1 = 2^kq$, ahol $q$ páratlan szám. Legyen $a$ egy $p$-vel nem osztható szám. Ekkor az alábbi két feltétel valamelyike teljesül: 
1. $a^q \equiv 1 \pmod{p}$
2. $a^q,a^{2q},a^{4q},\dots,a^{2^{k-1}q}$ valamelyike kongruense $-1 \mod{p}$-nek.

**Definíció**: Legyen $n$ páratlan szám és $n-1 = 2^kq$, ahol $q$ páratlan szám. Egy $a$ egész számot, amire $\text{gcd}(a,n)=1$, az $n$ Miller-Rabin tanújának nevezünk, ha az alábbiak teljesülnek:
1. $a^q \not\equiv 1 \pmod{n}$
2. $a^{2^iq} \not\equiv -1 \pmod{n}$, minden $i = 0, 1, 2, \dots, k-1$-re. 

Azaz ha létezik olyan $a$, ami Miller-Rabin tanúja $n$-nek, akkor $n$ biztosan összetett szám. Ebből felépíthető a Miller-Rabin teszt összetett számokra:

**Input**: $n$ egész szám, amit tesztelünk és $a$ potenciális tanú
1. Ha $n$ páros vagy $1 < \text{gcd}(a,n) < n$, akkor $n$ **összetett**.
2. $n - 1 = 2^kq$, ahol $q$ páratlan szám
3. Legyen $a = a^q \pmod{n}$
4. Ha $a \equiv 1 \pmod{n}$, adjuk vissza, hogy **a teszt elbukott**
5. Minden $i = 0, 1, 2, \dots, k - 1$-re:
    - Ha $a \equiv -1 \pmod{n}$, adjuk vissza, hogy **a teszt elbukott**
    - Legyen $a = a^2 \mod{n}$
6. Adjuk vissza, hogy **összetett**

Tehát generáljunk sok $a$ számot véletlenszerűen, majd nézzük meg a Miller-Rabin teszt eredményét. Segítségünkre lesz a következő tétel:

**Tétel**: Legyen $n$ páratlan összetett szám. Ekkor az $1 \leq a \leq n-1$ számok legalább 75%-a Miller-Rabin tanúja $n$-nek.

## Pollard $p-1$ algoritmusa

Legyenek $p,q$ prímszámok és $N=pq$ (csak $N$-et ismerjük). **Feladat**: határozzuk meg $p$-t és $q$-t!

Kemény munkával találtunk egy olyan $L$ egész számot úgy, hogy $p-1$ osztja $L$-et és $q-1$ nem osztja $L$-et. Ekkor léteznek olyan $i,j,k$ számok ($k \ne 0$), hogy $L = i(p-1)$ és $L = j(q-1)+k$. 

Vegyünk egy tetszőleges $a$ egész számot és számoljuk ki $a^L$ értékét:
\begin{align*}
    a^L &= a^{i(p-1)} = (a^{p-1})^i \equiv 1^i \equiv 1 \pmod{p} \\
    a^L &= a^{j(q-1)+k} = a^k (a^{q-1})^j \equiv a^k \cdot 1^j \equiv a^k \pmod{q}
\end{align*}

Mivel $p$ és $q$ nagyon nagyok és $k\ne 0$, elég nagy valószínűséggel $p$ osztani fogja $a^L -1$-et és $q$ nem fogja osztani $a^L -1$-et. Ebből $p$-t egyszerűen megkaphatjuk: $p = \text{gcd}(a^L - 1, N)$ és készen vagyunk.



Honnan fogunk ilyen $L$ számot találni? Pollard: ha $p-1$ kis prímszámok szorzataként előáll, akkor nem túl nagy $n$-re osztani fogja $n!$-t. A módszer: 
- vegyünk $n=2,3,\dots$ számokat és számoljuk ki $\text{gcd}(a^{n!}-1,N)$ értékét
    - gyakorlatban $a=2$
- ha a gcd $1$, akkor növeljük $n$-et egyel
- ha a gcd $N$, akkor válasszunk másik $a$-t
- ha $1 < \text{gcd} < N$, akkor $N$ egy nemtriviális felbontását kapjuk és készen vagyunk

**Tanulság**: $p-1$ és $q-1$ felbontásának köze van $N$ felbontásához! Válasszunk olyan $p$-t és $q$-t, aminél $p-1$ nem fog kis prímszámok szorzataként előállni.

```python
N = 13927189

def pollard_p(N):
    a = 2
    n = 2
    z = Zmod(N) # optimialization
    a = z(2)
    while True:
        temp = ((a ^ factorial(n)) - 1) % N # optimalization
        g = gcd(temp, N)

        if g == 1:
            n = n + 1
        elif g == N:
            a = a + 1
            n = 2
        else:
            return g


print(pollard_p(N))

# p = , q = 
```

### Pollard $\rho$ algoritmusa

Legyen $S$ egy véges halmaz és $f: S \rightarrow S$ egy függvény. Tegyük fel, hogy egy $x \in S$ elemre alkalmazzuk az $f$ függvényt egymás után, ami a következő sorozatot eredményezi:
$$
x_0 = x,\quad x_1 = f(x_0),\quad x_2 = f(x_1),\quad x_3 = f(x_2),\quad \dots,
$$
azaz $x_i = (\underbrace{f \circ f \circ f \circ \cdots \circ f}_\text{i-szer})(x)$

Mivel $S$ véges halmaz, előbb-utóbb lesz ismétlődés. Egy ideig $x_0,x_1,x_2,\dots$ egy irányba mozog, ahol az elemek nem ismétlődnek (ennek a hossza $T$ lesz). Egy idő után ($M$ lépés után) az elemek ismétlődni fognak.

![Pollard rho algoritmusa](./pollard_rho.png)

Legyen $S$ elemszáma $N$. Ekkor $T + M$ hossza nagyjából $1.2533 \sqrt{N}$. Mivel $T$ és $M$ értékét nem ismerjük előre, elő kell állítani az $x_0, x_1, x_2, \dots, x_{M + T}$ sorozatot.

(Egyező elemeket

Pollard: számoljunk két sorozatot
- a fenti $x_i$ sorozatot, és
- egy $y_i$ sorozatot úgy, hogy $y_0 = x_0$ és $y_{i+1} = f\left( f(y_i) \right)$ minden $i = 0,1,2,\dots$, azaz $y_i = x_{2i}$

Mikor találunk egyezést (*ütközést*)? Általánosan, $j > i$ esetén $x_j = x_i$ akkor és csak akkor, ha $i \geq T$ és $j \equiv i \pmod{M}$. Konkrétan: $x_{2i} = x_i$ akkor és csak akkor, ha $i \geq T$ és $2i \equiv i \pmod{M}$. Ez utóbbiból $M\;|\;i$, azaz akkor egyeznek, ha $i$ egyenlő lesz $M$ első többszörösével, ami $T$-nél nagyobb. 

#### Példa Pollard $\rho$ algoritmusára

**Prímfaktorizálás**: Legyenek $p,q$ $n$-bit hosszú prímszámok, $N = pq$ és $f : \mathbb{Z}_N^* \rightarrow \mathbb{Z}_N^*$. Válasszunk tetszőleges $x_0 \in \mathbb{Z}_N^*$ elemet és legyen $f(x) = (x^2 + 1 \mod{N})$. Ekkor a következő algoritmus előállítja $N$ egy prímtényezőjét:

1. $x_0 \leftarrow \mathbb{Z}_N^*$
2. $x' := x := x_0$
3. **for** $i = 1$ **to** $2^{n/2}$:
    - $x := f(x)$
    - $x' := f(f(x'))$
    - $p := \text{gcd}(x - x', N)$
    - **if** $p \not\in \{1, N\}$ **return** p és megállunk

**Példa**: Keressük meg $N = 37241$ prímtényezős felbontását!
