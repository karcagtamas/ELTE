```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

df = spark.read.json("./mongo_restaurant_dataset.json")
df.show()
```

```python
df.filter(df["borough"]=="Brooklyn").filter(df["cuisine"]=="Hamburgers").select(df["name"], df["address"]["street"]).show()
```

```python
df.filter(df["borough"]=="Brooklyn").groupBy(df["cuisine"]).count().show()
```
