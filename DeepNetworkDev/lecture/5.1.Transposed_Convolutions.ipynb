General formulas to know for this notebook

#### General formula for calculating the output size of a Convolution layer

Input size: $ c \times h \times w $

Filter: size $ c \times f \times f $ , stride $ s $ , padding $ p $ , number of filters $ m $


Output size: $ [ \; \frac{h + 2p - f}{s} + 1 \; ] \;\; \times \;\; [ \; \frac{w + 2p - f}{s} + 1 \; ] \;\; \times \;\; [ \; m \; ] $

#### General formula for calculating the output size of a Transposed Convolution layer

Input size: $ c \times h \times w $

Filter: size $ c \times f \times f $ , stride $ s $ , padding $ p $ , number of filters $ m $


Output size: $ [ \; (h - 1) * s - 2p + (f - 1) + 1 \; ] \;\; \times \;\; [ \; (w - 1) * s - 2p + (f - 1) + 1 \; ] \; \times \; [ \; m \; ] $

## 2. Transposed Convolution on 1D input

**Notes**

The **Transposed Convolutional Layer** is also (wrongfully) known as the Deconvolutional layer.
A deconvolutional layer reverses the operation of a standard convolutional layer i.e. if the output generated through a standard convolutional layer is deconvolved, you get back the original input. 

The transposed convolutional layer is similar to the deconvolutional layer in the sense that the spatial dimension generated by both are the same. But **Transposed convolution doesn’t reverse the standard convolution by values, rather by dimensions only**.

#### Input size: 2

#### Filter: size=3 , stride=2 , padding=0 , number of filters=1

![title](images/transconv1d1.png)

#### Formula for calculating the output size of a Transposed Convolution layer on 1D example

Input size: $ 2 $

Filter: $ f = 3 $ , $ s = 2 $ , $ p = 0 $ , number of filters $ 1 $


Output size: $ [ \; (h - 1) * s - 2p + (f - 1) + 1 \; ] \;\; \times \;\; [ \; (w - 1) * s - 2p + (f - 1) + 1 \; ] \; \times \; [ \; m \; ] $

Output size: $ (2 - 1) * 2 - 2*0 + (3 - 1) + 1  $

Output size: $ 1*2 - 0 + 2 + 1 $

Output size: $ 5 $

```python
#The conv1d will receive an input with 1 channel only and we will apply a single filter of size 3 with stride 2 and padding 0(valid).
convtrans1d = nn.ConvTranspose1d(in_channels=1, out_channels=1, kernel_size=3, stride=2, padding=1, bias=False)

convtrans1d.weight.data = torch.nn.parameter.Parameter(torch.LongTensor([[[2, 2, 2]]]), requires_grad=False)

print(f'Conv Transpose 1d weights {convtrans1d.weight}')
```

```python
with torch.no_grad():
    output = convtrans1d(input)
print("Output shape: ", output.shape)
print("Output values: ", output)
```

## Summary of Convolution vs Transposed Convolution on 1D input

![title](images/transconv1dpad9.png)

###### Remarks
Different from in the regular convolution where padding is applied to input, in the transposed convolution it is applied to output. For example, when specifying the padding number on either side of the height and width as 1, the first and last rows will be removed from the transposed convolution output.

In the transposed convolution, strides are specified for intermediate results (thus output), not for input.
